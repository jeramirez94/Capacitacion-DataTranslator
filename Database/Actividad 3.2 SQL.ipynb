```python
import numpy as np
import pandas as pd
import sqlite3 # <-- Interactuar con BD SQLite
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conn = sqlite3.connect("hr.sqlite"); conn
```

```python
#Mostrar los nombres (first_name, last_name) utilizando el alias “first name”, “last name” de la tabla de empleados
resultSet = pd.read_sql_query("select first_name as first_name, last_name as last_name from employees", conn);
resultSet
```

```python
#Obtener el “departmet ID” único de la tabla de empleados
sql = """
select distinct department_id
from   employees
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener todos los detalles de los empleados de la tabla de empleados ordenados por nombre, de manera descendente
resultSet = pd.read_sql_query("select employee_id, first_name, last_name, email, phone_number, hire_date, job_id, salary, commission_pct, manager_id, department_id, Avg_Salary from employees order by first_name desc", conn);
resultSet
```

```python
#Obtener nombre, apellido, salario y la remuneración de todos los empleados 
#(la remuneración se calcula como el 12% del salario).
sql = """
select first_name, last_name, salary * 0.12  remuneration, salary
from   employees
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el  “employee ID”, nombre, apellido y salario ordenados de manera ascendente por el salario.
resultSet = pd.read_sql_query("select employee_id, first_name, last_name, salary from employees order by salary asc", conn);
resultSet
```

```python
#Obtener el total del salario pagado a los empleados.
#-1
sql = """
select first_name, last_name, salary + commission_pct  total_salary
from   employees
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el salario máximo y mínimo de la tabla de empleados.
resultSet = pd.read_sql_query("select min(salary) minimun, max(salary) maximum from employees", conn);
resultSet
```

```python
#Obtener el salario medio y el número de empleados de la tabla de empleados.
sql = """
select avg(salary) avg_salary, count(employee_id) count_employee
from   employees
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el número de puestos de trabajo disponibles en la tabla de empleados.
resultSet = pd.read_sql_query("select count(distinct job_id) as available_positions from employees", conn);
resultSet
```

```python
#Obtener todos los nombres de la tabla de empleados en mayúsculas.
sql = """
select upper(first_name)
from   employees
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener los 3 primeros caracteres del nombre de cada empleado.
resultSet = pd.read_sql_query("select substr(first_name, 1, 3) from employees", conn);
resultSet
```

```python
#Obtener el nombre completo de todos los empleados de la tabla de empleados.
sql = """
select first_name || ' ' || last_name
from   employees
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el salario mensual (a 2 decimales) de todos los empleados.
resultSet = pd.read_sql_query("select round(salary,2) as salary from employees", conn);
resultSet
```

```python
#Obtener nombre, apellido y salario de todos los empleados cuyo salario no esté en el rango de 10,000 a 15,000.
sql = """
select first_name, last_name, salary
from   employees
where  salary < 10000 or salary > 15000
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener nombre, apellido y departamento de todos los empleados de los departamentos 
#30 o 100 en orden alfabético por “department ID”.
resultSet = pd.read_sql_query("select e.first_name, e.last_name, department_name "+
                              "from employees e " +
                              "join department d on e.department_id = d.department_id "+
                              "where e.department_id in (30, 100) " +
                              "order by department_name asc", conn);
resultSet
```

```python
#Obtener nombre, apellido y salario de todos los empleados cuyo salario 
#no esté en el rango de 10,000 a 15,000 y estén en el departamento 30 o 100.
sql = """
select first_name, last_name, salary
from   employees
where  salary < 10000 or salary > 15000
    and department_id in (30,100)
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el nombre de todos los empleados que tengan tanto una "b" y "c" en su nombre.
resultSet = pd.read_sql_query("select first_name from employees where first_name like '%b%c%' ", conn);
resultSet
```

```python
#Obtener el apellido, el cargo y el salario de todos los empleados 
#cuyo trabajo es el de “Programmer” o “Shipping Clerk”, y cuyo salario no sea igual a 4,500, 10,000 o 15,000 dólares.
sql = """
select e.last_name, j.job_title, e.salary
from   employees e
join   jobs j
   on  e.job_id = j.job_id
where  e.salary not in (4500, 10000, 15000)
   and j.job_title in ('Programmer', 'Shipping Clerk')
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener los apellidos de los empleados cuyos nombres tienen exactamente 6 caracteres.
resultSet = pd.read_sql_query("select last_name from employees where length(first_name) = 6 ", conn);
resultSet
```

```python
#Obtener los apellidos de los empleados que tienen "e" como tercer carácter.
sql = """
select e.last_name
from   employees e
where  substr(last_name, 3, 1) = 'e'
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el salario máximo de un empleado que trabaja como “Programmer”.
resultSet = pd.read_sql_query("select max(e.salary) max_salary_as_programmer "+
" from   employees e "+
" join   jobs j "+
   " on  e.job_id = j.job_id "+
" where  j.job_title = 'Programmer' ", conn);
resultSet
```

```python
#Obtener el salario medio y el número de empleados que trabajan en el departamento 90.
sql = """
select avg(e.salary) as avg_salary, count(employee_id)
from   employees e
where  e.department_id = 90
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el salario más alto, el más bajo, la suma y la media de todos los salarios.
resultSet = pd.read_sql_query("select max(salary) as max, min(salary) as min, sum(salary) as sum, avg(salary) as avg "+
                              " from employees ", conn);
resultSet
```

```python
#Obtener el número de empleados con el mismo puesto.
sql = """
select j.job_title, count(e.employee_id) as amount_employee
from   employees e
join   jobs j
   on  e.job_id = j.job_id
group by j.job_title
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener la diferencia entre el salario más alto y el más bajo.
resultSet = pd.read_sql_query("select max(salary) as max, min(salary) as min, max(salary) - min(salary) as difference"+
                              " from employees ", conn);
resultSet
```

```python
#Obtener el “manager ID” y el salario del empleado con menor sueldo reportando a ese “manager''.
sql = """
select e.manager_id, min(salary) min_salary
from   employees e
group by e.manager_id
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el “department ID” y el total de salarios a pagar en cada departamento.
resultSet = pd.read_sql_query("select e.department_id, sum(salary) total_salary, count(salary) amount_salaries "+
"from   employees e "+
"group by e.department_id", conn);
resultSet
```

```python
#Obtener el salario medio de cada ID de puesto, excluyendo “Programmer”.
sql = """
select j.job_id, avg(e.salary) as avg_salary_except_programmer
from   employees e
join   jobs j
   on  e.job_id = j.job_id
where  j.job_title <> 'Programmer'
group by j.job_id
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el job ID y el salario máximo de los empleados que su salario máximo es mayor o igual a $4,000.
resultSet = pd.read_sql_query("select e.job_id, max(salary) max_salary "+
"from   employees e "+
"group by e.job_id "+
"having max(salary) >= 4000", conn);
resultSet
```

```python
#Obtener el salario medio de todos los departamentos que emplean a más de 10 trabajadores.
sql = """
select d.depart_name, avg(e.salary) as avg_salary
from   employees e
join   departments d
   on  e.department_id = d.department_id
group by d.depart_name
having count(employee_id) > 10
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener nombre, apellido y  salario de los empleados que tienen un salario más alto 
#que el empleado cuyo apellido es “Bull”.

resultSet = pd.read_sql_query("select e.first_name, e.last_name, e.salary "+
"from   employees e "+
"where  salary > (select e2.salary from employees e2 where last_name = 'Bull') ", conn);
resultSet

```

```python
#Obtener nombre y apellido de todos los empleados que trabajan en el departamento de IT.
sql = """
select e.first_name, e.last_name
from   employees e
join   departments d
   on  e.department_id = d.department_id
where  d.depart_name = 'IT'
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener nombre y apellido de los empleados que tienen un “manager” que trabaja en un departamento con sede en EUA.
resultSet = pd.read_sql_query("select e.first_name, e.last_name, m.first_name "+
                              "from   employees e "+
                              "join employees m "+
                              " on e.manager_id = m.employee_id "+
                              "join departments d on m.department_id = d.department_id "+
                              "join locations l on d.location_id = l.location_id "+
                              "join countries c on l.country_id = c.country_id "+
                              "where c.country_id = 'US'", conn);
resultSet
```

```python
#Obtener nombre y apellido de los empleados que son “managers”.
sql = """
select distinct m.first_name, m.last_name
from   employees e
join   employees m
  on e.manager_id = m.employee_id
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener nombre, apellido y salario de los empleados cuyo salario es mayor que el salario medio.
resultSet = pd.read_sql_query("select e.first_name, e.last_name, e.salary "+
                            "from   employees e "+
                            "where  salary > (select avg(e2.salary) from employees e2) ", conn);
resultSet
```

```python
#Obtener nombre, apellido y salario de los empleados cuyo salario es igual al salario mínimo para su rol.
sql = """
select e.first_name, e.last_name
from   employees e
join   jobs j
    on e.job_id = j.job_id
    and e.salary = j.min_salary
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener nombre, apellido y salario de los empleados que ganan más que el 
#salario medio y que trabajan en cualquier área de IT.
resultSet = pd.read_sql_query("select e.first_name, e.last_name, e.salary, department_name "+
                            "from   employees e "+
                            "join department d "+
                              " on e.department_id = d.department_id "
                            "where  salary > (select avg(e2.salary) from employees e2) "+
                              "and d.department_name like 'IT%'", conn);
resultSet
```

```python
#Obtener nombre, apellido y salario de los empleados que ganan más que el Sr. Bell.
sql = """
select e.first_name, e.last_name, e.salary
from   employees e
where  salary > (select e2.salary from employees e2 where last_name = 'Bell')
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener nombre y apellido de los empleados que no son “supervisors”.
sql = """
select e1.first_name, e1.last_name
from   employees e1
where  e1.employee_id not in
                (   select distinct m.employee_id
                    from   employees e
                    join   employees m
                      on e.manager_id = m.employee_id
                )
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el employee ID, nombre, apellido y salario de todos los empleados 
#cuyo salario es superior a la media de sus departamentos.

sql = """
select e1.first_name, e1.last_name, e1.last_name
from   employees e1
inner join (
    select d.department_id, avg(e.salary) as avg_salary
    from   employees e
    join   departments d
    on  e.department_id = d.department_id
    group by d.depart_name
) dep
ON  e1.department_id = dep.department_id
WHERE   e1.salary > dep.Avg_Salary
"""
pd.read_sql_query(sql, conn)
```

```python
#Obtener el 5º salario máximo de la tabla de empleados.
resultSet = pd.read_sql_query("select first_name, last_name, salary FROM ("+
                             "select e.first_name, e.last_name, e.salary "+
                             "from   employees e "+
                             "order by salary desc "+
                              "Limit 5) " + 
                              " order by salary asc LIMIT 1", conn);
resultSet
```

```python
#Obtener las direcciones (location_id, street_address, city, state_province, country_name) de todos los departamentos.
resultSet = pd.read_sql_query("select d.depart_name, d.location_id, l.street_address, l.city, state_province, c.country_name "+
                              "from departments d  "+
                              "join locations l on d.location_id = l.location_id "+
                              "join countries c on l.country_id = c.country_id "+
                              "where c.country_id = 'US'", conn);
resultSet
```

```python
#Obtener el “employee ID”, nombre y apellido de cada empleado junto con el nombre, apellido y “manager ID” de su “manager”.
sql = """
select e.employee_id, e.first_name, e.last_name, m.first_name as first_name_manager, m.last_name as last_name_manager
from   employees e
join   employees m
  on e.manager_id = m.employee_id
"""
pd.read_sql_query(sql, conn)
```
