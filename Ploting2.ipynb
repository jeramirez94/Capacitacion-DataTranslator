```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import default_rng

%matplotlib inline
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(12,12))
ax= fig.add_subplot(111,projection="3d")
for c,z in zip(["red","green", "blue","yellow"],[0,10,20,30]):
    xs=np.arange(20)
    ys=np.random.rand(20)
    cs=[c]*len(xs)
    cs[0]="orange"
    ax.bar(xs,ys,zs=z, zdir="y", color=cs, alpha=0.7)
    
a=ax.set_xlabel("X")
b=ax.set_ylabel("Y")
c=ax.set_zlabel("Z")
```

```python
fig=plt.figure(figsize=(6,6))
with plt.style.context("dark_background"):
    N=4
    x=np.linspace(0,N)
    ncolors=len(plt.rcParams["axes.prop_cycle"])
    shift=np.linspace(0,N, ncolors,endpoint=False)
    for l in shift:
        plt.plot(x, np.sin(x+l),'--')
```

```python
import seaborn as sns
x=np.arange(0,4,0.1)
y=np.sin(x)
df=pd.DataFrame({'x':x,'sinx':y,'cos':np.cos(x)})
df=df.set_index('x')
grid=sns.lmplot(x='x',y='sinx',data=df.reset_index())
grid.axes[0,0].hlines(0,-1,5) #mover los ejes
ax=grid.axes
```

```python
sns.set(style="dark")
rs=np.random.RandomState(4)
pos = rs.randint(-1,2,(20,5)).cumsum(axis=1)
step = np.tile(range(5),20)
walk = np.repeat(range(20),5)
df = pd.DataFrame(np.c_[pos.flat, step, walk], columns=['pos','step','walk'])

grid=sns.FacetGrid(df, col='walk',hue='walk',col_wrap=5)
#lnea en Y=0
grid.map(plt.axhline, y=0, ls=':')
#Lnea con o
grid.map(plt.plot, 'step','pos', marker='o', ms=4)
```
